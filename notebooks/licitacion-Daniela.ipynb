```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df["diferencia_amount"]=  df["value_amount"] - df["amount"]
df["diferencia_amount2"]=  df["amount"] - df["value_amount"]
df["diferencia_amount"]
```

¿Cuáles fueron los métodos de licitación que generaron la mayor diferencia de contrato a nivel nacional y por provincias? 

```python
metodo_licitacion_prov = df.groupby(["Provincia", "procurementMethodDetails"])["diferencia_amount"].agg('sum').reset_index()
#metodo_licitacion_nac = df.groupby("procurementMethodDetails").agg({'diferencia_amount': ['sum', 'mean', 'min', 'max','size']}).reset_index()
metodo_licitacion_nac = df.groupby("procurementMethodDetails")["diferencia_amount"].agg('sum').reset_index()

metodo_licitacion_nac
```

```python
maximo_difmont = metodo_licitacion_nac.diferencia_amount.max()
maximo_difmont
```

```python
min_difmont = metodo_licitacion_nac.diferencia_amount.min()
min_difmont
```

¿Cuál es la frecuencia más alta o representativa del monto del contrato teniendo en cuenta la duración en días de este? 

```python
df_frec = df[["contractPeriod_durationInDays","diferencia_amount", "diferencia_amount2"]]
df_frec
```

```python
df_describe_monto = df["diferencia_amount"].describe()
minimo = df_describe_monto["min"]
primer_cuartil = df_describe_monto["25%"]
media = df_describe_monto["50%"]
maximo = df_describe_monto["max"]
```

```python
df_describe_dias = df["contractPeriod_durationInDays"].describe()
minimo_dias = df_describe_dias["min"]
primer_cuartil_dias = df_describe_dias["25%"]
media_dias = df_describe_dias["50%"]
maximo_dias = df_describe_dias["max"]
```

```python
df_describe_dias
```

```python
df_frec['range_monto'] = pd.cut(df_frec["diferencia_amount"], [minimo,primer_cuartil,media,maximo], include_lowest=True)
```

```python
df_frec['range_dias'] = pd.cut(df_frec["contractPeriod_durationInDays"], [minimo_dias,media_dias,maximo_dias], include_lowest=True)
```

```python
df_frec['range_monto']
```

```python
table= pd.pivot_table(df_frec, index=['range_dias'], columns='range_monto', aggfunc=np.size,values="diferencia_amount")
table2= pd.pivot_table(df_frec, index=['range_dias'], columns='range_monto', aggfunc=np.size,values="contractPeriod_durationInDays")
```
