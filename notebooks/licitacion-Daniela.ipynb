```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv("../data/dataset-RedFlags.csv")
df
```

```python
df_contract_2018 = pd.read_csv("../data/releases_2018/contracts_2018.csv")
df_contract_2019 = pd.read_csv("../data/releases_2019/contracts_2019.csv")
df_contract_2020 = pd.read_csv("../data/releases_2020/contracts_2020.csv")
df_contract_2021 = pd.read_csv("../data/releases_2021/contracts_2021.csv")
df_contract = pd.concat([df_contract_2018, df_contract_2019, df_contract_2020, df_contract_2021], axis=0)
```

```python
df_award_2018 = pd.read_csv("../data/releases_2018/awards_2018.csv")
df_award_2019 = pd.read_csv("../data/releases_2019/awards_2019.csv")
df_award_2020 = pd.read_csv("../data/releases_2020/awards_2020.csv")
df_award_2021 = pd.read_csv("../data/releases_2021/awards_2021.csv")
df_award = pd.concat([df_award_2018, df_award_2019, df_award_2020, df_award_2021], axis=0)
```

```python
df_tender_2018 = pd.read_csv("../data/releases_2018/tender_2018.csv")
df_tender_2019 = pd.read_csv("../data/releases_2019/tender_2019.csv")
df_tender_2020 = pd.read_csv("../data/releases_2020/tender_2020.csv")
df_tender_2021 = pd.read_csv("../data/releases_2021/tender_2021.csv")
df_tender = pd.concat([df_tender_2018, df_tender_2019, df_tender_2020, df_tender_2021], axis=0)
```

```python
filt_tender = df_tender[["ocid", "procuringEntity_name", "procuringEntity_id", "procurementMethod"]]
filt_award = df_award[["ocid", "id", "amount",]]
filt_contract = df_contract[["ocid", "id", "amount", "contractPeriod_durationInDays"]]
```

```python
df_tender.dtypes
```

```python
df_contract.dtypes
```

```python
preg1 = filt_award.merge(filt_contract, left_on="ocid", right_on="ocid", how="left", suffixes=("_aw", "_ctr"))
preg1 = preg1[~(preg1["amount_ctr"].isnull()) & ~(preg1["amount_aw"].isnull())]
preg1 = preg1.merge(filt_tender, left_on="ocid", right_on="ocid", how="left")
preg1
```

```python
provincias = pd.read_csv("../data/provinciasRUC.csv")
```

```python
preg1['ruc']=preg1['procuringEntity_id'].str.split('-').str[2]
```

```python
preg1['codigoProvincia']=preg1['ruc'].str[:2].astype(int)
preg1 = preg1.merge(provincias,left_on = "codigoProvincia", right_on="Codigo", how="left")
```

```python
preg1.drop(columns=["codigoProvincia","Codigo"], inplace=True)
```

```python
preg1["diferencia_absoluta"] = np.abs(preg1["amount_aw"] - preg1["amount_ctr"])
preg1.head()
```

¿Cuáles fueron los métodos de adquisición que generaron la mayor diferencia de contrato a nivel nacional y por provincias? 

```python
metodo_licitacion_prov = preg1.groupby(["Provincia", "procurementMethod"])["diferencia_absoluta"].agg('sum').reset_index().sort_values(by=["diferencia_absoluta"],ascending=False)
#metodo_licitacion_nac = df.groupby("procurementMethodDetails").agg({'diferencia_amount': ['sum', 'mean', 'min', 'max','size']}).reset_index()
metodo_licitacion_nac = preg1.groupby("procurementMethod")["diferencia_absoluta"].agg('sum').reset_index()
```

```python
metodo_licitacion_nac=metodo_licitacion_nac.sort_values(by=["diferencia_absoluta"],ascending=False)
metodo_licitacion_nac
```

```python
maximo_difmont = metodo_licitacion_nac.diferencia_absoluta.max()
maximo_difmont
```

```python
min_difmont = metodo_licitacion_nac.diferencia_absoluta.min()
min_difmont
```

```python
table_prov= pd.pivot_table(metodo_licitacion_prov, index='Provincia', columns=['procurementMethod'], aggfunc= ["sum"],values="diferencia_absoluta")
```

```python
table_prov.dropna()
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
ax = sns.barplot(y='diferencia_absoluta', x='procurementMethod',data=metodo_licitacion_nac)
ax.title.set_text('Métodos de adquisición que generaron la mayor diferencia absoluta entre precio adjudicado y precio declarado de contrato a nivel nacional\n')
ax.set_ylabel('Diferencia de cantidad del contrato')
ax.set_xlabel('Métodos de adquisición')
ax.ticklabel_format(useOffset=False, style='plain', axis= 'y')
plt.savefig("preg9_nac.png", bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
ax = sns.heatmap(table_prov,cmap='YlGnBu')
ax.title.set_text('Métodos de adquisición que generaron la mayor diferencia absoluta entre precio adjudicado y precio declarado de contrato por provincia\n')
ax.set_xlabel('Métodos de adquisición')
ax.set_ylabel('Provincia')
plt.show()
plt.savefig("preg9_prov.png", bbox_inches='tight')
```

```python
#sns.barplot(x='name',y='totalCount',data=metodo_licitacion_nac)
fig,axs = plt.subplots(nrows=2,ncols=1,figsize=(15,30))
ax = sns.barplot(y='diferencia_absoluta', x='procurementMethod', ax=axs[0],data=metodo_licitacion_nac)
ax.title.set_text('Métodos de adquisición que generaron la mayor diferencia absoluta entre precio adjudicado y precio declarado de contrato a nivel nacional\n')
ax.set_ylabel('Diferencia de cantidad del contrato')
ax.set_xlabel('Métodos de adquisición')
ax.ticklabel_format(useOffset=False, style='plain', axis= 'y')
#ax.set_xlim(min_difmont, 0)
ax = sns.heatmap(table_prov,ax=axs[1],cmap='YlGnBu')
ax.title.set_text('Métodos de adquisición que generaron la mayor diferencia absoluta entre precio adjudicado y precio declarado de contrato por provincia\n')
ax.set_xlabel('Métodos de adquisición')
ax.set_ylabel('Provincia')
plt.show()
```

¿Cuál es la frecuencia más alta o representativa del monto del contrato teniendo en cuenta la duración en días de este? 

```python
df_frec = preg1[["contractPeriod_durationInDays","amount_ctr"]]
df_frec
```

```python
df_frec["amount_ctr"].unique()
```

```python
df_frec["range_monto"] = pd.cut(df_frec["amount_ctr"], bins=10)
df_frec["range_dias"] = pd.cut(df_frec["contractPeriod_durationInDays"], bins=5)
```

```python
df_frec["range_monto"].unique()
```

```python
table= pd.pivot_table(df_frec, index=['range_dias'], columns='range_monto', aggfunc=np.size,values="amount_ctr")
```

```python
table=table.fillna(0)
table
```

```python

plt.figure(figsize = (10,7))
plt.title('Frecuencia del monto según la duración del contrato en días \n')
sns.heatmap(table,cmap='YlGnBu')
plt.xlabel('Monto')
plt.ylabel('Duración de contrato en días')
plt.savefig("preg10.png", bbox_inches='tight')
```
